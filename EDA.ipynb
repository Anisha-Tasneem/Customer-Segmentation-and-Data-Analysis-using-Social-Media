```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your CSV file
# Replace 'your_file.csv' with the actual file name or path
file_path = '/content/drive/MyDrive/Cse499/OurData.csv'
customer_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(customer_data.head())

# Summary statistics
print("\nSummary statistics:")
print(customer_data.describe())

# Check for missing values
print("\nMissing values:")
print(customer_data.isnull().sum())

# Distribution of categorical variables
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.countplot(x='Gender', data=customer_data)
plt.title('Distribution of Gender')

plt.subplot(2, 2, 2)
sns.countplot(x='City', data=customer_data)
plt.title('Distribution of City')

plt.subplot(2, 2, 3)
sns.countplot(x='Product Details (Brand)', data=customer_data)
plt.title('Distribution of Product Brand')

plt.subplot(2, 2, 4)
sns.countplot(x='Address', data=customer_data)
plt.title('Distribution of Address')

plt.tight_layout()
plt.show()

# Distribution of numerical variables
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
sns.histplot(customer_data['Product Price (Taka)'], bins=20, kde=True)
plt.title('Distribution of Product Price')

plt.subplot(1, 3, 2)
sns.histplot(customer_data['Devivery Charge (Taka)'], bins=20, kde=True)
plt.title('Distribution of Delivery Charge')

plt.subplot(1, 3, 3)
sns.histplot(customer_data['Total Bill(Taka)'], bins=20, kde=True)
plt.title('Distribution of Total Bill')

plt.tight_layout()
plt.show()
```
